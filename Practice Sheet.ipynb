```python
#       0  1   2  3  4  5  6  7
nums = [5, -2, 3, 9, 0, 6, 10, 7]
k = 5
n = len(nums)

k = k % n

def reverse(nums, left, right):
    while left < right:
        nums[left], nums[right] = nums[right], nums[left]
        left += 1
        right -= 1


reverse(nums, n - k, n - 1) #[      7,10,6,0,9]

reverse(nums, 0, n - k - 1) #[3,-2,5          ]

reverse(nums, 0, n - 1)     #[9,0,6,10,7,5,-2,3]

print(nums)
```
