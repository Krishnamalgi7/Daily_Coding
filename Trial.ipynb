```python
nums = [1, 1, 2, 3, 3, 4, 4, 5, 5, 6, 7, 8, 9, 10, 11, 11]

n = len(nums)

freq_map = {}

for i in range(0, n):
    freq_map[nums[i]] = 0

j = 0

for k in freq_map:
    nums[j] = k
    j = j + 1
    print(k, end=" ")
```
